```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.array([0,6,12,18,24])
y = np.array([7,149,23,88,35])
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'o')
plt.show()
```

```python
plt.plot(y)
```

```python
plt.plot(x,y,'o-.r')
```

```python
plt.plot(y,x,'o:b',ms='10')
```

```python
plt.plot(y,x,'X:y',ms='10',mec='b')
```

```python
##line style
plt.plot(x,y,linestyle='dotted')
```

```python
plt.plot(x,y,ls='-.')
```

```python
##labels
plt.plot(x,y)
plt.xlabel("time")
plt.ylabel("speed")
```
