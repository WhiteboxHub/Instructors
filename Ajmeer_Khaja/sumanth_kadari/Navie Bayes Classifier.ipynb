```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

dataset=pd.read_csv("C:\\Users\\hr\\Downloads\\Social_Network_Ads.csv")
dataset.head()
```

```python
#checking empty values 
dataset.isna().sum()
```

```python
#creatwe features matrix and dependant variable vector
x= dataset.iloc[:,[2,3]].values #independt values
y= dataset.iloc[:,4].values #target values/dependt
x[0]
```

```python
#splitting data into training set and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x, y,test_size=0.2, random_state=0)
```

```python
#feature scalling 
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
#traing the model
#fitting naive bayes to the training set
#traning data will be continous values
#gaussian is wrk for mostly continoues values
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(x_train, y_train)
```

```python
#predicting the test set results
y_pred= classifier.predict(x_test)
y_pred
```

```python
#confusion matrix /here we pred how much our model wkrs using cm
#it will be represnts as true positive,negative and false,positve,negative
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test, y_pred)
import  seaborn as sns
sns.heatmap(cm,annot=True)
```
