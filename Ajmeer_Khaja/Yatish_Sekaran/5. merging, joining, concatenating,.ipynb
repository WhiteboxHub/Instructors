# Merging by Indexes with the left_index and right_index Parameters

```python
week1 = pd.read_csv(r"C:\Users\vvick\Downloads\Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv(r"C:\Users\vvick\Downloads\Restaurant - Week 2 Sales.csv")
customers = pd.read_csv(r"C:\Users\vvick\Downloads\Restaurant - Customers.csv", index_col='ID')
food = pd.read_csv(r"C:\Users\vvick\Downloads\Restaurant - Foods.csv", index_col='Food ID')
```

```python
customers.head()
```

```python
food.head()
```

```python
week1.head()
```

```python
sales = week1.merge(customers, how = 'left', left_on = 'Customer ID', right_index = True)
sales
```

```python
sales.merge(food, how = 'left', left_index = True, right_on='Food ID')
```

```python
week2.merge(customers, how = 'left', right_on = 'ID', left_index = True, )
```
