```python
import matplotlib.pyplot as plt

xp1= [1,2,3]
yp1 =[7,7,7]
xp2= [2,2,1,2]
yp2 =[7,3,4,5]
xp3= [3,2,3,3,2]
yp3 =[7,5,5,3,3]
plt.figure(figsize=(5,6))
plt.title("FMT")
plt.plot(xp1,yp1,'*:g')
plt.plot(xp2,yp2,'.:b')
plt.plot(xp3,yp3,'.:')
plt.show()
```

```python
import matplotlib.pyplot as plt

xp1= [1,2,3]
yp1 =[7,7,7]
xp2= [2,2,1,2]
yp2 =[7,3,4,5]
xp3= [3,2,3,3,2]
yp3 =[7,5,5,3,3]
plt.figure(figsize=(5,6))
plt.title("FMT using mec and mfc")
plt.plot(xp1,yp1,'*--g',ms=10,mec='black',mfc='orange')
plt.plot(xp2,yp2,'.--b',ms=15,mec='black',mfc='orange')
plt.plot(xp3,yp3,'.--',ms=20,mec='black',mfc='orange')
plt.show()
```

```python
import matplotlib.pyplot as plt

x1=[0,5,10,15]
x2=[5,10,15]
y1=[0,10,20,30]
y2=[0,20,40]

font1 = {'family':'Times new roman','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Two plots", loc ="left",fontdict = font1)
plt.xlabel("x_axis", fontdict = font1)
plt.ylabel("Y_axis", fontdict = font2)
plt.plot(x1,y1,"orange")
plt.plot(x2,y2,"r")
plt.legend(["Speed1","Speed2"])
plt.grid()
plt.show()
```
