# Filter a DataFrame's Rows with String Methods

```python
employees = pd.read_csv(r"C:\Users\vvick\Downloads\employees.csv").dropna()
employees.head()
```

```python
employees['Team']
```

```python
mask = employees['Team'].str.lower().str.contains('services')
mask
```

```python
employees[mask]
```

```python
mask2 = employees['Team'].str.lower().str.startswith('client')
employees[mask2]
```

```python
mask3 = employees['Team'].str.lower().str.endswith('ing')
employees[mask3]
```
