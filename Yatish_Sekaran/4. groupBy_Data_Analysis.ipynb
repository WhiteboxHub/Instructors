# The .agg() Method

```python
fortune = pd.read_csv(r"C:\Users\vvick\Downloads\Fortune_1000.csv", usecols=['company','rank','revenue','sector','profit'])
sectors = fortune.groupby('sector')
fortune.head()
```

```python
sectors.agg({'revenue' : ['sum','max'],
            'profit' : ['mean','sum']})
```

```python
sectors.agg(['sum','size'])
```

# Iterating Through Groups

```python
fortune = pd.read_csv(r"C:\Users\vvick\Downloads\Fortune_1000.csv")
sectors = fortune.groupby('sector')
fortune.head()
```

```python
df = pd.DataFrame(columns = fortune.columns)
df
```

```python
for sector, data in sectors:
    highest_revenue_company_in_group = data.nlargest(1, 'revenue')
    df = df._append(highest_revenue_company_in_group)
```

```python
df
```

```python
cities = fortune.groupby('city')
```

```python
df = pd.DataFrame(columns = fortune.columns)
df
```

```python
for city, data in cities:
    highest_revenue_by_city = data.nlargest(1, 'revenue')
    df = df._append(highest_revenue_by_city)
```
