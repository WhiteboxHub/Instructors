```python
#Central Limit Theorem


import numpy as np
import matplotlib.pyplot as plt

# Parameters
population_mean = 100  # Mean of the population
population_std = 20     # Standard deviation of the population
sample_size = 30        # Sample size
num_samples = 1000      # Number of samples to draw

# Generate random samples from the population
population = np.random.normal(population_mean, population_std, 10000)

# Calculate sample means
sample_means = [np.mean(np.random.choice(population, sample_size)) for _ in range(num_samples)]

# Plot histogram of sample means
plt.figure(figsize=(10, 6))
plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='b', edgecolor='black')

# Plot the theoretical normal distribution based on Central Limit Theorem
sample_mean_std = population_std / np.sqrt(sample_size)
x = np.linspace(min(sample_means), max(sample_means), 100)
y = 1/(sample_mean_std * np.sqrt(2 * np.pi)) * np.exp( - (x - population_mean)**2 / (2 * sample_mean_std**2) )
plt.plot(x, y, color='red', linewidth=2)

plt.title('Distribution of Sample Means vs. Theoretical Normal Distribution')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.legend(['Theoretical Normal Distribution', 'Sample Means'])
plt.grid(True)
plt.show()
```

```python
#Standard Normal Distribution


import numpy as np
import matplotlib.pyplot as plt

# Parameters
sample_size = 30        # Sample size
num_samples = 1000      # Number of samples to draw
exponential_rate = 0.5  # Rate parameter for exponential distribution

# Generate random samples from the exponential distribution
population = np.random.exponential(1/exponential_rate, 10000)

# Calculate sample means
sample_means = [np.mean(np.random.choice(population, sample_size)) for _ in range(num_samples)]

# Plot histogram of sample means
plt.figure(figsize=(10, 6))
plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='b', edgecolor='black')

# Plot the theoretical normal distribution based on Central Limit Theorem
sample_mean_std = 1 / (exponential_rate * np.sqrt(sample_size))  # Standard deviation of sample means for exponential distribution
x = np.linspace(min(sample_means), max(sample_means), 100)
y = 1/(sample_mean_std * np.sqrt(2 * np.pi)) * np.exp( - (x - 1/exponential_rate)**2 / (2 * sample_mean_std**2) )
plt.plot(x, y, color='red', linewidth=2)

plt.title('Distribution of Sample Means from Exponential Distribution vs. Theoretical Normal Distribution')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.legend(['Theoretical Normal Distribution', 'Sample Means'])
plt.grid(True)
plt.show()
```

```python
#68-95-99.5 Rule

import numpy as np
import matplotlib.pyplot as plt

# Generate random samples from a normal distribution
np.random.seed(0)  # For reproducibility
data = np.random.normal(loc=0, scale=1, size=10000)

# Calculate mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

# Define the ranges
one_std_range = (mean - std_dev, mean + std_dev)
two_std_range = (mean - 2*std_dev, mean + 2*std_dev)
three_std_range = (mean - 3*std_dev, mean + 3*std_dev)

# Count the percentage of data within each range
within_one_std = np.mean(np.logical_and(data >= one_std_range[0], data <= one_std_range[1])) * 100
within_two_std = np.mean(np.logical_and(data >= two_std_range[0], data <= two_std_range[1])) * 100
within_three_std = np.mean(np.logical_and(data >= three_std_range[0], data <= three_std_range[1])) * 100

# Print the results
print("Percentage of data within one standard deviation:", within_one_std)
print("Percentage of data within two standard deviations:", within_two_std)
print("Percentage of data within three standard deviations:", within_three_std)

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, density=True, alpha=0.6, color='b', edgecolor='black')

# Plot the ranges
plt.axvline(one_std_range[0], color='red', linestyle='--')
plt.axvline(one_std_range[1], color='red', linestyle='--')
plt.axvline(two_std_range[0], color='green', linestyle='--')
plt.axvline(two_std_range[1], color='green', linestyle='--')
plt.axvline(three_std_range[0], color='purple', linestyle='--')
plt.axvline(three_std_range[1], color='purple', linestyle='--')

# Add labels and legend
plt.title('Histogram of Random Data with Standard Deviation Intervals')
plt.xlabel('Data')
plt.ylabel('Density')
plt.legend(['Data', 'One Std Dev', 'Two Std Dev', 'Three Std Dev'])
plt.grid(True)
plt.show()
```

```python
#The Beta Distribution

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Parameters
alpha = 2   # Shape parameter
beta_val = 5  # Shape parameter
sample_size = 30        # Sample size
num_samples = 1000      # Number of samples to draw

# Generate random samples from the Beta distribution
population = beta.rvs(alpha, beta_val, size=10000)

# Calculate sample means
sample_means = [np.mean(np.random.choice(population, sample_size)) for _ in range(num_samples)]

# Plot histogram of sample means
plt.figure(figsize=(10, 6))
plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='b', edgecolor='black')

# Plot the theoretical normal distribution based on Central Limit Theorem
sample_mean_std = np.std(population) / np.sqrt(sample_size)  # Standard deviation of sample means for Beta distribution
x = np.linspace(min(sample_means), max(sample_means), 100)
y = 1/(sample_mean_std * np.sqrt(2 * np.pi)) * np.exp( - (x - np.mean(population))**2 / (2 * sample_mean_std**2) )
plt.plot(x, y, color='red', linewidth=2)

plt.title('Distribution of Sample Means from Beta Distribution vs. Theoretical Normal Distribution')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.legend(['Theoretical Normal Distribution', 'Sample Means'])
plt.grid(True)
plt.show()
```

```python
#Binomial Distribution

import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 20  # Number of trials
p = 0.5  # Probability of success
num_samples = 1000  # Number of samples to draw

# Generate random samples from the binomial distribution
population = np.random.binomial(n, p, size=10000)

# Calculate sample means
sample_means = [np.mean(np.random.choice(population, n)) for _ in range(num_samples)]

# Plot histogram of sample means
plt.figure(figsize=(10, 6))
plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='b', edgecolor='black')

# Calculate the mean and standard deviation of the population
population_mean = np.mean(population)
population_std = np.std(population)

# Plot the theoretical normal distribution based on Central Limit Theorem
sample_mean_std = population_std / np.sqrt(n)  # Standard deviation of sample means for binomial distribution
x = np.linspace(min(sample_means), max(sample_means), 100)
y = 1/(sample_mean_std * np.sqrt(2 * np.pi)) * np.exp( - (x - population_mean)**2 / (2 * sample_mean_std**2) )
plt.plot(x, y, color='red', linewidth=2)

plt.title('Distribution of Sample Means from Binomial Distribution vs. Theoretical Normal Distribution')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.legend(['Theoretical Normal Distribution', 'Sample Means'])
plt.grid(True)
plt.show()

```
